```python
#compare target productivity with actual

import altair as alt
import pandas as pd
data = pd.read_csv("gar_data.csv")

base = alt.Chart(data).encode(x='month(date):O')
bar = base.mark_bar().encode(y='actual_productivity:Q')

line =  base.mark_line(color='red').encode(
    y='targeted_productivity:Q'
)
(bar + line).properties(width=600)
```

```python
#to determine the pattern for productivity to explore the relationship between departmetnt, overtime, no of workers and actual productivity
import altair as alt
import pandas as pd
data = pd.read_csv("gar_data.csv")
alt.Chart(data).mark_area(opacity=0.3).encode(
    x="quarter:O",
    y=alt.Y("actual_productivity:Q", stack=None),
    color="department:N"
).properties(height=250, width=900)

```

The layered histogram shows the productivity of departments across the three months. As expected fininshing is more productive in the final stage of production.  it also indicates the finishing departmetnt 

```python
#to determine the pattern for productivity to explore the relationship between departmetnt, overtime, no of workers and actual productivity
import altair as alt
import pandas as pd
data = pd.read_csv("gar_data.csv")
alt.Chart(data).properties(width=550)

base =line = base.mark_line().encode(
    x='month(date):O',
    y='no_of_workers',
    color='average(actual_productivity)'
)

rule = base.mark_rule().encode(
    y='average(targeted_productivity)',
    color='department',
    size=alt.value(2)
)

line + rule
```
